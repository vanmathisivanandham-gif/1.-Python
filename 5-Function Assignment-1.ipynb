2. Create a function that checks whether the given number is Odd or Even

```python
def oddeven():
        num=int(input("Enter a number:"))
        if (num%2==1):
            print(num, "is Odd number")
            message="odd number"
        else:
            print(num, "is Even number")
            message="Even number"
        return message
```
