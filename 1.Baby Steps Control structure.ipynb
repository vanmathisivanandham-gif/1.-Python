3.Body Mass index

```python
BMIRange=int(input("Enter the BMI Index"))
if(BMIRange<16):
    print("Severly underweight")
elif(BMIRange<18):
    print("Underweight")
elif(BMIRange<25):
    print("Normal")
elif(BMIRange<30):
    print("Overweight")
else:
    print("Very Overweight")
    
```
